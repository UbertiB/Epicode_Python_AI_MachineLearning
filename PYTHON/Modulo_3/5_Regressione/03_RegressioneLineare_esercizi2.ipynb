Utilizzando il dataset Diabetes, replicare l'esercitazione pratica in Python
- Caricare il dataset
- Addestrare un oggetto di regressione lineare e ricavare le y_pred
- Creare scatter plot dei data point target
- Creare scatter plot dei data point reali e predetti differenziandoli per colore
- Creare scatter plot dei risultati (MAE) con riferimento all'asse y=0 (errore nullo)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
```

Carico il dataset

```python
data=load_diabetes(as_frame=True) #carica direttamente il dataframe pandas
df=data.frame
print(f"Feature: {data.feature_names}")
print(df.head())
X=df[data.feature_names]
y=df["target"]
```

- Addestrare un oggetto di regressione lineare e ricavare le y_pred

Creo il modello

```python
model=LinearRegression()
model.fit(X,y)
```

Valore predetti

```python
y_pred=model.predict(X)
```

- Creare scatter plot dei data point target

```python
plt.figure(figsize=(8,6))
plt.scatter(y,y_pred,alpha=0.3)
plt.plot([y.min(),y.max()],[y.min(),y.max()],color="red",linestyle="--")
plt.xlabel("Valori reali")
plt.ylabel("Valori predetti")
plt.title("Predizione valore medio delle abitazioni in California")
plt.grid()
plt.show()
```

Tutto quello che è sopra la linea rosso è correttamente predetto, tutto quello che c'è sopra è sovrastimato, quello che c'è sotto è sottostimato

- Creare scatter plot dei data point reali e predetti differenziandoli per colore

```python
plt.figure(figsize=(8,6))
# punti reali
plt.scatter(y, y, color="blue", alpha=0.5, label="Reali")
# punti predetti
plt.scatter(y, y_pred, color="orange", alpha=0.5, label="Predetti")
# linea ideale
plt.plot([y.min(), y.max()], [y.min(), y.max()],
         color="red", linestyle="--", label="Perfetto")
plt.xlabel("Valori reali")
plt.ylabel("Valori")
plt.title("Reale vs Predetto")
plt.legend()
plt.grid()

plt.show()
```

- Creare scatter plot dei risultati (MAE) con riferimento all'asse y=0 (errore nullo)

```python
error=y.squeeze()-y_pred.ravel()  #MAE
plt.figure(figsize=(8,6))
plt.scatter(y,error,alpha=0.6,s=50,label="Target")
plt.axhline(y=0,color="red",linestyle="--",linewidth=2,label="Errore nullo")
plt.xlabel("Target")
plt.ylabel("Errore")
plt.title("Distribuzione errori")
plt.legend()
plt.grid()
plt.show()
```
