Utilizzando il dataset Iris, replicare l'esercitazione pratica in Python.
    - Caricare il dataset
    - Addestrare un oggetto regressione logistica (max_iter=5_000)
    - Fare predizioni del target e calcolare il numero di etichette corrette e accuratezza
    - Creare scatter plot data point reali e predetti defferenziandoli per colore
    - Creare scatter plot data point Petal Lenght (sull'asse X), reali e predetti (sull'asse Y), differenziandoli per colore.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

- Caricare il dataset

```python
data=load_iris()
#print(data.DESCR)
print(data.feature_names)
print(len(data.feature_names))
```

```python
X,y=data.data, data.target
print(y)
df=pd.DataFrame(X,columns=data.feature_names)
print(df.head())
df["target"]=y
df.dtypes
```

- Addestrare un oggetto regressione logistica (max_iter=5_000)

```python
model=LogisticRegression(max_iter=5_000)
model.fit(X,y)
```

- Fare predizioni del target e calcolare il numero di etichette corrette e accuratezza

```python
print(f"Intercetta: {model.intercept_[0]}")
print(f"Coefficienti principali:")
for name,coef in zip(data.feature_names[::],model.coef_[0][::]):
    print(f"{name}:{coef:.3f}")
```

```python
y_pred=model.predict(X)
```

```python
corrette=np.sum(y==y_pred)
tot=len(y)
accurancy=corrette/tot
print(f"Etichette corrette: {corrette}/{tot}")
print(f"Accuracy: {accurancy*100:.2f}")
```

- Creare scatter plot data point reali e predetti defferenziandoli per colore

```python
plt.figure(figsize=(8,6))
plt.scatter(range(tot),y,label="Reali",alpha=0.6)
plt.scatter(range(tot),y_pred,label="Predette",alpha=0.6)
plt.title("Grafico Reali-Predette")
plt.xlabel("Indice campione")
plt.ylabel("Classe")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

- Creare scatter plot data point Petal Lenght (sull'asse X), reali e predetti (sull'asse Y), differenziandoli per colore.

```python
plt.figure(figsize=(8,6))
plt.scatter(df["petal length (cm)"],y,label="Reali",alpha=0.4)
plt.scatter(df["petal length (cm)"],y_pred,label="Predetti",alpha=0.4, color="red")
plt.title("Relazione Mean radius - Classe")
plt.xlabel("Mean radius")
plt.ylabel("Class")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
