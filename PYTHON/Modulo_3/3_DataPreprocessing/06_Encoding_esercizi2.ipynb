Il CODING di una variabile categorica serve a trasformare etichette testuali o classi in numeri che un modello di ml posso poi usare.
Un algoritmo non 'capisce' davvero il significato delle etichette passate (esempio rosso, blu, verde). Capisce solo i numeri. Ma nei numeri c'è una trappola: non basta sostituire le parole con numeri qualsiasi (rosso=1, blue=2, verde=3), perchè questo fa credere al modello che il valore di un etichetta sia maggiore/minore rispetto ad un'altra verde>blu>rosso. Cioè introduce un ordine che non esiste. Per questo di utilizzano tecniche di encoding.

1) Variabile categorica
Una variabile categorica contiene classi.
Esempi: 1) rosso, blu verde 2) reparto amministrazione, reparto logistica, reparto vendite 3) cliente retail, cliente grossista, cliente estero.

Ci sono due grandi famiglie:
1) NOMINALI: nessun ordine naturale
   Esempio: colore
2) ORDINALI: c'è un ordine
   Esempio: basso, medio, alto
One-hot encoding, dummy ed effect coding si usa soprattutto per i nominali

2) Non basta assegnare un numero
Perchè essendo variabili nominali non c'è un ordine tra i vari valori, nell'assegnare un numero il modello potrebbe imparare relazioni che non esistono.

ONE-HO CODING: il più intuitivo
Per ogni categoria crei una colonna nuova, e metti 1 se la riga appartiene a quella categoria, 0 altrimenti
rosso=[1,0,0]
blu=[0,1,0]
verde=[0,0,1]
In questo modo non si introduce un ordine falso.
Difetto: se hai molte categorie, il numero di colonne esplode.

DUMMY CODING
Quasi uguale a one-hot coding ma con una categoria di riferimento.
Togli  una colonna rispetto a one-hot, perchè la categoria mancante diventa la categoria di riferimento
Esempio verde=riferimento
rosso=[1,0]
blu=[0,1]
verde=[0,0]
Si utilizza per evitare la multicollinearità perfetta nei modelli lineari.
Se usi tutte le colonne (come in one-hot) una colonna è ricavabile dalle altre, il modello ha ridondanza.

EFFECT CODING
Stessa idea del dummy, ma il riferimento non vale 
Nell'effect coding usi ancora k-1 colonne per k categorie, ma la categoria di riferimento viene codificata con -1 invece che con 0
Esempio verde=riferimento
rosso=[1,0]
blu=[0,1]
verde=[-1,-1]
Con dummy coding, i coefficienti si leggono come differenza rispetto alla categoria di riferimento
Con effect coding, i coefficienti si leggono come differenza rispetto alla media complessiva delle categorie.

Riassunto secco:
One-hot: una colonna per categoria
         Rappresenta ogni categoria separatamente
         Molto intuitivo, Ottimo per preprocessing generico
Dummy: una colonna in meno, riferimento = tutti 0
       Rappresenta ogni categoria rispetto a una categoria base.
       Perfetto per regressione e modelli lineari interpretabili
Effect: una colonna in meno, riferimento = tutti -1
        Rappresenta ogni categoria rispetto alla media complessiva.
        Utile quando vuoi interpretazioni statistiche più raffinate

IL PROBLEMA DELLE TANTE CATEGORIE
Se hai una variabile come codice_articolo, cliente, fornitore, con migliaia di valori one-hot diventa spetto ingestibile.
A questo punto si valutano altre tecniche:
- target coding
- frequency encoding
- hasing encoding
- embedding
