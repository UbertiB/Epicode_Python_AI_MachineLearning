Recall

```python
from sklearn.metrics import recall_score
```

Classificazione Binaria

```python
y_true=["A"]*9+["B"]*6
y_pred=["A"]*7+["B"]*2+["B"]*4+["A"]*2
recall_manuale_a=7/(7+2)
print (f"Recall calcolata manualmente - Classe A: {recall_manuale_a*100:.2f}")

recall_a=recall_score(y_true,y_pred,pos_label="A")
recall_b=recall_score(y_true,y_pred,pos_label="B")
print (f"Precison calcolato con Python - Classe A: {recall_a*100:.2f}")
print (f"Precison calcolato con Python - Classe B: {recall_b*100:.2f}")
```

Classificazione multiclasse

```python
y_true=["A","A","A","B","B","B","C","C","C","A","C","B"]
y_pred=["C","C","A","B","C","B","A","A","A","A","B","B"]

recall_micro=recall_score(y_true,y_pred,average="micro")
recall_macro=recall_score(y_true,y_pred,average="macro")
recall_weighted=recall_score(y_true,y_pred,average="weighted")

print (f"Recall micro: {recall_micro*100:.2f}%")
print (f"Recall macro: {recall_macro*100:.2f}%")
print (f"Recall weighted: {recall_weighted*100:.2f}%")
```
