Accuracy binaria

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score
```

Considerando un dataset di 15 esempi, in cui:
- 9 appartengono alla classe A
- 6 appartengono alla classe B
Ipotizzando che le predizioni del modello siano:
TP=7, TN=4, FP=2, FN=2
Calcolare manualmente l'accuracy
Verificare il risultato usando i metodi di Python

```python
print(F"Accuracy calcolato manualmente:{(7+4)/(7+4+2+2)*100:.2f}%")
```

```python
y_true = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B']
y_pred = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'A', 'A']
```

```python
accuracy = accuracy_score(y_true, y_pred)
```

```python
print(f"Accuracy calcolato con metodo Python: {accuracy*100:.2f}%")
```

Esercizio: Classificazione Multiclasse

```python
y_true=["A","A","A","B","B","B","C","C","C","A","B","C"]
y_pred=["A","B","A","B","C","B","C","A","C","C","B","B"]
accuracy=balanced_accuracy_score(y_true,y_pred)
print(f"Accuracy balancy calcolato con metodo Python: {accuracy*100:.2f}%")
```
