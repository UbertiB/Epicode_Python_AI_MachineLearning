Utilizzando il dataset California_Housing, replicare l'esercitazione pratica
- Effettuare la standarizzazione dei dati
- Addestarre tra modelli SVR, per ognuno calcolare MSE e R2
    - kernel="linear", C=1.0 eps=5.0
    - kernel="rbf", C=1.0 eps=5.0
    - kernel="rbf", C=1.0 eps=20.0
- Scegliere una feature per effettuare il plot di regressione (x: feature y:target)
    - scatter plot dei dati reali
    - le tra rette di regressione con le relative bande di eps

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data=fetch_california_housing()
X=data.data
y=data.target
```

- Effettuare la standarizzazione dei dati

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

Addestrare 3 modelli

linear

```python
svr_linear=SVR(kernel="linear",C=1.0,epsilon=5.0)
svr_linear.fit(X_scaled,y)
y_pred_linear=svr_linear.predict(X_scaled)
mse_linear=mean_squared_error(y,y_pred_linear)
r2_linear=r2_score(y,y_pred_linear)
```

rbf eps=5.0

```python
svr_rbf_narrow=SVR(kernel="rbf",C=1.0,epsilon=5.0)
svr_rbf_narrow.fit(X_scaled,y)
y_pred_narrow=svr_rbf_narrow.predict(X_scaled)
mse_narrow=mean_squared_error(y,y_pred_narrow)
r2_narrow=r2_score(y,y_pred_narrow)
```

rbf eps=20.0

```python
svr_rbf_wide=SVR(kernel="rbf",C=1.0,epsilon=20.0)
svr_rbf_wide.fit(X_scaled,y)
y_pred_wide=svr_rbf_wide.predict(X_scaled)
mse_wide=mean_squared_error(y,y_pred_wide)
r2_wide=r2_score(y,y_pred_wide)
```

```python
print(f"LINEAR")
print(f"\tMSE: {mse_linear}\t R2:{r2_linear}")
print(f"RBF (eps=5)")
print(f"\tMSE: {mse_narrow}\t R2:{r2_narrow}")
print(f"RBF (eps=20)")
print(f"\tMSE: {mse_wide}\t R2:{r2_wide}")
```

Le metriche MSE e R2 sono sostanzialmente identiche per i 3 modelli addestrati, questo vuol dire che eps è stato impostato in modo errato ed il modello non fitta. Per impostare eps è necessario controllare i valori di y
Inoltre R2 negativo tutti i valori predetti potrebbero essere uguali.
il tutto è validato anche dai seguenti print

```python
print(y.min(), y.max(), y.mean())
print(y_pred_linear.min(), y_pred_linear.max(), y_pred_linear.mean())
print(y_pred_narrow.min(), y_pred_narrow.max(), y_pred_narrow.mean())
print(y_pred_wide.min(), y_pred_wide.max(), y_pred_wide.mean())
```

pertanto cambio i valori di eps e ricalcolo tutto

```python
svr_linear=SVR(kernel="linear",C=1.0,epsilon=0.1)
svr_linear.fit(X_scaled,y)
y_pred_linear=svr_linear.predict(X_scaled)
mse_linear=mean_squared_error(y,y_pred_linear)
r2_linear=r2_score(y,y_pred_linear)
```

```python
svr_rbf_narrow=SVR(kernel="rbf",C=1.0,epsilon=0.1)
svr_rbf_narrow.fit(X_scaled,y)
y_pred_narrow=svr_rbf_narrow.predict(X_scaled)
mse_narrow=mean_squared_error(y,y_pred_narrow)
r2_narrow=r2_score(y,y_pred_narrow)
```

```python
svr_rbf_wide=SVR(kernel="rbf",C=1.0,epsilon=0.5)
svr_rbf_wide.fit(X_scaled,y)
y_pred_wide=svr_rbf_wide.predict(X_scaled)
mse_wide=mean_squared_error(y,y_pred_wide)
r2_wide=r2_score(y,y_pred_wide)
```

```python
print(y.min(), y.max(), y.mean())
print(y_pred_linear.min(), y_pred_linear.max(), y_pred_linear.mean())
print(y_pred_narrow.min(), y_pred_narrow.max(), y_pred_narrow.mean())
print(y_pred_wide.min(), y_pred_wide.max(), y_pred_wide.mean())
```

```python
print(f"LINEAR")
print(f"\tMSE: {mse_linear}\t R2:{r2_linear}")
print(f"RBF (eps=0.1)")
print(f"\tMSE: {mse_narrow}\t R2:{r2_narrow}")
print(f"RBF (eps=0.5)")
print(f"\tMSE: {mse_wide}\t R2:{r2_wide}")
```

Linear non fitta bene R2=0.085 vuol dire che linear sta spiegando solo 8.5% della variabilità del target. In pratica cattura pochissimo. La relazione tra feature e target non è ben rappresentata da un confine lineare.
I modelli MSE invece vanno decisamente meglio con circa 75% (r2). Quindi sicuramente la relazione tra X  y non è lineare.
Per RBF e eps 0.1 e 0.5 i valori sono quasi identici, quindi nel range provato di eps, il comportamento del modello è sostanzialmente identico.
Quindi scartando linear ora mi concentro su RBF, provo diversi valori per vedere il cambiamento

```python
for eps in [0.01, 0.1, 0.5]:
    for C in [1, 5, 10]:
        for gamma in ["scale", 0.1, 1]:
            model = SVR(kernel="rbf", epsilon=eps, C=C, gamma=gamma)
            model.fit(X_scaled, y)
            y_pred = model.predict(X_scaled)
            print(
                f"eps={eps}, C={C}, gamma={gamma} | "
                f"MSE={mean_squared_error(y, y_pred):.4f} | "
                f"R2={r2_score(y, y_pred):.4f}"
            )
```
