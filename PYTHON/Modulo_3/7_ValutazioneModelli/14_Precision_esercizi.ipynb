PRECISION

```python
from sklearn.metrics import precision_score
```

Classificazione Binaria

```python
y_true=["A"]*9+["B"]*6
y_pred=["A"]*7+["B"]*2+["B"]*4+["A"]*2
precision_manuale_a=7/(7+2)
print (f"Precision calcolata manualmente - Classe A: {precision_manuale_a*100:.2f}")

precision_a=precision_score(y_true,y_pred,pos_label="A")
precision_b=precision_score(y_true,y_pred,pos_label="B")
print (f"Precison calcolato con Python - Classe A: {precision_a*100:.2f}")
print (f"Precison calcolato con Python - Classe B: {precision_b*100:.2f}")
```

Classificazione multiclasse

```python
y_true=["A","A","A","B","B","B","C","C","C","A","C","B"]
y_pred=["C","C","A","B","C","B","A","A","A","A","B","B"]

precision_micro=precision_score(y_true,y_pred,average="micro")
precision_macro=precision_score(y_true,y_pred,average="macro")
precision_weighted=precision_score(y_true,y_pred,average="weighted")

print (f"Precisoni micro: {precision_micro*100:.2f}%")
print (f"Precisoni macro: {precision_macro*100:.2f}%")
print (f"Precisoni weighted: {precision_weighted*100:.2f}%")
```
